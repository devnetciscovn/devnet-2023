LIST MESSAGES

```python
import requests
from webex_credentials import base_url, token, room_id_phuc

url = f'{base_url}/v1/messages?roomId={room_id_phuc}'
headers = {
    'Authorization': f'Bearer {token}'
} 

#print(url)
#print(headers)

response = requests.get(url, headers=headers)
response.status_code
```

```python
response.json()
```

```python
response.json()['items']
```

```python
for item in response.json()['items']:
    print(item)
```

```python
for item in reversed(response.json()['items']):
    print(item['personEmail'], item['text'], sep='\n\t')
```

QUERY PARAMETERS

```python
import requests
from webex_credentials import base_url, token, room_id_phuc
```

```python
url = f'{base_url}/v1/messages'
headers = { 
    'Authorization': f'Bearer {token}'
}
params = {
    'roomId': room_id_phuc
}

response = requests.get(url, headers=headers, params=params)

print(response.status_code)
```

CREATE A MESSAGE

```python
import requests
from webex_credentials import base_url, token, room_id_phuc

url = f'{base_url}/v1/messages'
headers = {
    'Authorization': f'Bearer {token}'
}
body = {
    "toPersonEmail": "huygiap67@gmail.com",
    "text": "Chào Giáp, đây là tin nhắn gửi từ Python"
}

response = requests.post(url, headers=headers, json=body)
response.status_code
```

DELETE A ROOM

```python
import requests
from webex_credentials import base_url, token, room_id_group_101

url = f'{base_url}/v1/rooms/{room_id_group_101}'

headers = {
    'Authorization': f'Bearer {token}'
}

response = requests.delete(url, headers=headers)
response.status_code
```
